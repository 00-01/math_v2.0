```python

t = turtle.Turtle()
t.getscreen().bgcolor("black")
x1, x0, xi, y1, y0, yi = 100, 0, -100, 100, 0, -100

def draw_background():
    t.speed(11)
    t.color("gray", "black")
    t.pensize(1)

  ## large square
    t.penup()
    t.goto(x1, y1)
    t.pendown()
    t.goto(xi, y1)
    t.goto(xi, yi)
    t.goto(x1, yi)
    t.goto(x1, y1)

    ## diagonal
    t.penup()
    t.goto(x1, y1)
    t.pendown()
    t.goto(xi, yi)
    t.penup()
    t.goto(xi, y1)
    t.pendown()
    t.goto(x1, yi)

    ##small square
    t.penup()
    t.goto(x1, y0)
    t.pendown()
    t.goto(x0, y1)
    t.goto(xi, y0)
    t.goto(x0, yi)
    t.goto(x1, y0)

    ## cross
    t.penup()
    t.goto(x1, y0)
    t.pendown()
    t.goto(xi, y0)
    t.penup()
    t.goto(x0, y1)
    t.pendown()
    t.goto(x0, yi)

def draw_line():
    t.speed(5)
    t.color("red", "black")
    t.pensize(3)

    ## large square
    t.color("green")
    t.penup()
    t.goto(x1, y1)
    t.pendown()
    t.goto(xi, y1)
    t.goto(xi, yi)
    t.goto(x1, yi)
    t.goto(x1, y1)

    ## diagonal
    t.color("yellow")
    t.penup()
    t.goto(x1, y1)
    t.pendown()
    t.goto(xi, yi)
    t.penup()
    t.goto(xi, y1)
    t.pendown()
    t.goto(x1, yi)

    ##small square
    t.color("orange")
    t.penup()
    t.goto(x1, y0)
    t.pendown()
    t.goto(x0, y1)
    t.goto(xi, y0)
    t.goto(x0, yi)
    t.goto(x1, y0)

    ## cross
    t.color("red")
    t.penup()
    t.goto(x1, y0)
    t.pendown()
    t.goto(xi, y0)
    t.penup()
    t.goto(x0, y1)
    t.pendown()
    t.goto(x0, yi)

def draw_dot():
    dot = 10
    t.ht()
    t.speed(5)
    t.color("blue")
    t.pensize(3)

    t.penup()
    t.goto(x1, y1)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(x0, y1)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(xi, y1)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(xi, y0)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(x0, y0)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(x1, y0)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(x1, yi)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(x0, yi)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(xi, yi)
    t.pendown()
    t.dot(dot)
# half
    t.penup()
    t.goto(x1/2, y1/2)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(xi/2, y1/2)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(xi/2, yi/2)
    t.pendown()
    t.dot(dot)

    t.penup()
    t.goto(x1/2, yi/2)
    t.pendown()
    t.dot(dot)
    # t.write("GFG", align="center", font=('Verdana', 12, 'bold'))



draw_background()
# draw_line()
draw_dot()


turtle.done()
turtle.exitonclick()

# make all possible num, alphabet
# have to be easily distinguishable

```
